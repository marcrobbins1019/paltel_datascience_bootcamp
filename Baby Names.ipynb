```python
def starting_with(letter, data = baby_names):
    return baby_names.loc[baby_names['Name'].str[0] == letter].groupby('Year').sum()['Count']
```

```python
starting_with('Z')
```

```python
for letter in ['A','B','C','D','E']:
    names = starting_with(letter)
    plt.plot(names, label = letter)
    
plt.legend()
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
    
```
